```python
def caeser_cipher(str,key1,key2):
    """_summary_
    encypt string with 2 key Caeser algorithm 
    Args:
        str (string): string to encrypt
        key1 (int): key1
        key2 (int): key2

    Returns:
        cstr (string): crypted string
    """
    key = [key1,key2]
    table = [chr(i) for i in range(ord('a'), ord('z') + 1)]
    tablec = [chr(i) for i in range(ord('A'), ord('Z') + 1)]
    punctuation = [' ',',','.','?','!',';','\"']
    cstr = ""
    index = 0
    for i in str:
        if i not in punctuation:
            if i in table:
                cstr += table[(table.index(i)+key[index%2])%26]
            if i in tablec:
                cstr += tablec[(tablec.index(i)+key[index%2])%26]
        else:
            cstr += i
        index += 1
        
    return cstr
                
        
sentence = "Aal uttx hm aal Qtct Fhljha pl Wbdl. Pvxvxlx!"

print(caeser_cipher(sentence,19,7))
```

```python
def de_crypt(str):
    table = [chr(i) for i in range(ord('a'), ord('z') + 1)]
    tablec = [chr(i) for i in range(ord('A'), ord('Z') + 1)]
    decrypt_str=[]
    punctuation = [' ',',','.','?','!',';','\"']
    for i in range(26):
        for j in range(26):
            index = 0
            temp_str = ""
            for c in str:
                if c not in punctuation:
                    ktable = [i,j]
                    if c in table:
                        temp_str += table[(table.index(c)+ktable[index%2])%26]
                    if c in tablec:
                        temp_str += tablec[(tablec.index(c)+ktable[index%2])%26]
                else:
                    temp_str += c
                index += 1
            decrypt_str.append(temp_str)
    
    return decrypt_str

encrypt_str = "Aal uttx hm aal Qtct Fhljha pl Wbdl. Pvxvxlx!"
index=0
re = de_crypt(encrypt_str)
fea = [0 for i in range(26*26)]
vowl = {'A':1,'E':1,'I':1,'O':1,'U':1,'a':1,'e':1, 'i':1, 'o':1, 'u':1}
for sentence in re:
    for c in sentence:
        try:
            fea[re.index(sentence)] += vowl[c]
        except KeyError:
            continue

print(max(fea),fea.index(max(fea)))
print(re[fea.index(max(fea))])
for i in re:
    print(i)
```
